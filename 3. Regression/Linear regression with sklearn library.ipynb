```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfx = pd.read_csv('linearX.csv')
dfy = pd.read_csv('linearY.csv')

x = dfx.values
y = dfy.values

x = x.reshape((-1,1))
y = y.reshape((-1,1))

X = (x-x.mean())/x.std()
Y = y
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression()
```

# Training

```python
model.fit(X,Y)
```

# Predictions

```python
output = model.predict(X)
```

# Parameters Learnedm

```python
bias = model.intercept_
coeff = model.coef_

print(bias)
print(coeff)
```

# Score

```python
model.score(X,Y)
```

```python
## Visualise

plt.scatter(X,Y,label='data')
plt.plot(X,output,color='orange',label='prediction')
plt.legend()
plt.show()
```
